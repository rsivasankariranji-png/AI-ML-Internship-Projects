```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
data = pd.read_csv("milk_demand.csv")

X = data[['Day']]
y = data['Milk_Sold']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict next 5 days
future = [[16],[17],[18],[19],[20]]
pred = model.predict(future)

print("Future Milk Demand Prediction:")
for i, v in enumerate(pred):
    print(f"Day {16+i} : {int(v)} liters")

# Plot graph
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Day")
plt.ylabel("Milk Sold")
plt.title("Milk Demand Prediction")
plt.show()
```
